```python
#import the libraries and input the data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
##extracting Independent and dependent variable
X = train_df.drop('label', axis= 1)
Y = train_df['label']
```

```python
#model training
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state=100)
```

```python
#data preprocessing
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

#fit transform use to do some calculation and do the transformation
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
##model construction
from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
rbf_model.fit(X_train, Y_train)
```

```python
#model testing
y_rbf_pred = rbf_model.predict(X_test)
```

```python
print("Predicted values :\n", y_rbf_pred[10:15])
print("Actual Values :\n", Y_test[10:15])
```

```python
from sklearn import metrics
acc_rbf = metrics.accuracy_score(Y_test , y_rbf_pred)
print("accuracy: ", "{:2f}".format(acc_rbf)*100, "%")
```

```python
#using random integers for testing
for i in (np.random.randint(0,270,4)):
    two_d = (np.reshape(X_test[i], (28,28)))
    plt.title('predicted label :{0}'.format(y_rbf_pred[i]))
    plt.imshow(two_d, cmap='gray')
    plt.show()
```
